```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato_DA.csv")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
def handelrate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handelrate) 
```

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("type of res")
```

```python
df.head()
```

```python
grp = df.groupby('listed_in(type)')['votes'].sum()
res = pd.DataFrame({'votes' : grp})
plt.plot(res)
```

```python
plt.hist(df['rate'] , bins=5)
plt.show()
```

```python
# tol = df['approx_cost(for two people)']
sns.countplot(x= df['approx_cost(for two people)'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x= 'online_order'  , y = 'rate'  , data=df)
plt.show()
```

```python
piv_tbl = df.pivot_table(index='listed_in(type)' , columns='online_order' , aggfunc='size' , fill_value=0)
sns.heatmap(piv_tbl , annot=True , cmap='YlGnBu' , fmt='d')
plt.show()
```
